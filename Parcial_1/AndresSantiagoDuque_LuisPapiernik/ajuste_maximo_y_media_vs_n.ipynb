```python
import numpy as np
import matplotlib.pyplot as plt
```

En la carpeta "files" se encuentran los archivos con los datos para cada *n* de la última ejecución del ejecutable del script "game_of_life", en el cual se realiza el ajuste a una gausiana con ayuda de la clase TH1 de root.

El archivo "Medias_maximos.txt" contiene la media y el maximo del número de ciclos que tardo en estabilizar el juego de la vida.

<h2> Previsualización de los datos</h2>

```python
data = np.loadtxt('./files/Medias_Maximos.txt')
```

```python
data.shape
```

```python
plt.plot(data[:,0],data[:,1])
plt.title("n vs maximo")
plt.xlabel("n")
plt.ylabel("maximo")
plt.show()
```

```python
plt.plot(data[:,0],data[:,2])
plt.title("n vs media")
plt.xlabel("n")
plt.ylabel("media")
plt.show()
```

Eliminando el último dato de la  gráfica "n vs máximo", el cúal es un dato atípico si se compara con los datos de la media, se obtiene el siguiente conjunto de datos que aparentan una relación lineal

```python
plt.plot(data[:54,0],data[:54,1])
plt.title("n vs maximo")
plt.xlabel("n")
plt.ylabel("maximo")
plt.show()
```

<h2>Ajuste de los datos</h2>

- *n vs máximo*

```python
p = np.poly1d(np.polyfit(data[:54,0],data[:54,1],1))

plt.plot(data[:54,0],data[:54,1],'*')
plt.plot(data[:54,0],p(data[:54,0]))
plt.title("n vs maximo")
plt.xlabel("n")
plt.ylabel("maximo")
plt.show()

print("La recta que mejor se ajusta a los datos es \ny = %.3f x +%.3f"%(p[0],p[1]))
```

- *n vs media

En este caso, no es tan claro la función a la cual se ajustan los datos. a continuación se realiza un ajuste lineal

```python
p = np.poly1d(np.polyfit(data[:54,0],data[:54,2],1))

plt.plot(data[:54,0],data[:54,2],'*')
plt.plot(data[:54,0],p(data[:54,0]))
plt.title("n vs media")
plt.xlabel("n")
plt.ylabel("media")
plt.show()

print("La recta que mejor se ajusta a los datos es \ny = %.3f x +%.3f"%(p[0],p[1]))
```
