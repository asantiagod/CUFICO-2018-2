```python
import ROOT
from array import array
from numpy.random import normal
from ROOT import gROOT 
```

```python
MyTree = ROOT.TTree( 'tree1', 'tree para un par de dados' )
```

```python
x = array( 'f', [ 0 ] )
y = array( 'f', [ 0 ] )
z = array( 'f', [ 0 ] )
```

```python
MyTree.Branch( 'x', x, 'x/F' )
MyTree.Branch( 'y', y, 'y/F' )
MyTree.Branch( 'z', y, 'z/F' )
```

```python
for i in range(10000):
  x[0] = normal(5,2)    
  y[0] = normal(5,2)
  z[0] = normal(5,2)    
  MyTree.Fill()
```

```python
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
```

<h3>1) x*y</h3>

```python
MyTree.Draw("x*y>>h1(100,0,100)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

<h3>2) x*z</h3>

```python
MyTree.Draw("x*z>>h2(100,0,100)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

<h3>3) x*y*z</h3>

```python
MyTree.Draw("x*y*z>>h3(100,0,1000)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

<h3>4) x si z entre 4.5 y 5.5</h3>

```python
MyTree.Draw("x>>h4(20,0,10)","z>=4.5 && z<=5.5")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

<h3>5) x:z  si y entre 4.5 y 5.5</h3>

```python
MyTree.Draw("x:z>>h5(20,0,10,20,0,10)","y>=4.5 && y<=5.5","colz")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

<h3>6) x:y:z</h3>

```python
MyTree.Draw("x:y:z>>h6(20,0,10,20,0,10,20,0,10)","","LEGO")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```
